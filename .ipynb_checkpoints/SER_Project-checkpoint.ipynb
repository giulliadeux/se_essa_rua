### 1. Overview and Background



#### 1.1 Introduction

<span style="font-family:Georgia;">While Brazil is internationally known for its rich culture - encompassing sport, food, dance, music, and so on - there’s very little acknowledgment that for many Brazilians, access to culture is still a privilege. Nearly a third of Brazil's kids don’t have access to culture; Overall, a large part of the population depends on free access to go to any cultural events.[1]</span>

<span style="font-family:Georgia;">In nations where access to culture is still not democratised, the streets are usually a good indicator of the communities expression of culture. Growing up in São Paulo, I witnessed culture manifest itself through vivid graffiti adorning every wall, painted goal-marks on the pavement, small garage improvised restaurants and remnants of the green and yellow decor from the most recent World Cup. The only aspect of a street that cannot be changed by the people living in it, is its name, an unchangeable feature shaped by external factors. 

<span style="font-family:Georgia;">Therefore, this study aims to explore and quantify the history of every street name in São Paulo, Brazil - a study referred to before with the term <strong>‘streetonomic’</strong> - a form of computational social science that studies human behavior and cultural trends through the quantitative analysis of street names.</span>

#### 1.2 Background


<span style="font-family:Georgia;">The idea behind this study comes from an enigmatic expression of Brazilian culture that has resonated with me since childhood.<strong> “Se essa rua fosse minha”</strong> is a traditional popular Brazilian song, estimated to be over a century old. Its lyrics are a powerful anthem, up for public interpretation, as there’s no explicit author. To me, this song encapsulates the desire to take ownership of our streets, not driven by power or vanity, but rather by the desire to improve the lives of the people we care about.</span>

### 5.1.2 Setup for Status Log

<span style="font-family:Georgia;"> Added a function to get the 'status' of the database every time a modification is made. This was proven necessary because I found myself having to display the size of the database every time I ran a new function to double check if the code worked.

```python
# defining first sizes of the database
first_size_DIC = DIC_database.shape[0]
first_size_CEM = CEM_database.shape[0]

# function to get the status of row changes after each operation
def rows_change_log(df, initial_size, first_size):
    new_size = df.shape[0]
    rows_removed = initial_size - new_size
    print(divider)
    print('[b black]STATUS[/b black]')
    print(f'Initial size of the database: [b black]{first_size}[/b black]')
    print(f'Number of rows before previous operation: [b black]{initial_size}[/b black]')
    print(f'Number of rows removed in previous operation: [b #e81a3c]{rows_removed}[/b #e81a3c]')
    print(f'New size of the database: [b green]{new_size}[/b green]')
    print(divider)

    return new_size
```

#### 5.1.4.2 Creating unique IDs

<span style="font-family:Georgia;">I decided to create unique IDs for each row in the databases motivated by the need to reference both the <code>street name</code> and the <code>history</code> data in the DIC database to identify duplicates in an efficient way, since reading the whole database and each string every time would be quite computationally expensive. The same is true for the CEM database, taking in consideration <code>street name</code> and the <code>km</code> data.
    
<span style="font-family:Georgia;">That been said, the unique IDs will serve as a distinguishing factor in the database, mainly for detection of duplicates. Example of the proposed format for the unique IDs:
    
> <span style="font-family:Georgia;"> <strong>DIC UID</strong> </span>
> <br> <span style="font-family:Georgia;"> 4 first letters of the street name + *dash* + 4 first letters of history, when available</span>

```python
def generate_uid(string, mode='initials'):
    if pd.isna(string): # if 'NaN', return an empty string
        return ''
    elif mode == 'initials':
        return ''.join(word[0] for word in string.split())[:4].lower()
    elif mode == 'join_words':
        return string.replace(' ', '').lower()

# creates DIC uid
DIC_database_clean['uid'] = (
    DIC_database_clean['street name'].apply(generate_uid, mode='join_words') +
    '-' +
    DIC_database_clean['history'].apply(generate_uid, mode='initials')
)


# creates CEM uid
CEM_database_clean['uid'] = (
    CEM_database_clean['street name'].apply(generate_uid, mode='join_words')
)


display(DIC_database_clean.head())
display(CEM_database_clean.head())
```

#### 5.1.4.3 Create Merging Rules

<span style="font-family:Georgia;"> We need to create merging rules that will be used on the merging operation; they exist so we don't break any other part of the database with this opperation. There are two different dictionaries for DIC and CEM because each database has a specific need.</span>

```python
# setup to start merging

def merge_or_join_comma(x):
    unique_values = x.dropna().unique()
    return unique_values[0] if len(unique_values) == 1 else ', '.join(map(str, unique_values))

def merge_or_join_underscore(x):
    unique_values = x.dropna().unique()
    return unique_values[0] if len(unique_values) == 1 else '_'.join(map(str, unique_values))


merging_rules_DIC = {
    'id': merge_or_join_underscore, # if ids are same return it, else join with _
    'district': merge_or_join_comma, # if districts are same return it, else join with ,
    'date of legislation': merge_or_join_comma, # if dates are same return it, else join with ,
    'full street name': merge_or_join_comma, # if full street names are same return it, else join with ,

}

merging_rules_CEM = {
    'id': merge_or_join_underscore, # if ids are same return one else join with _
    'city': merge_or_join_comma, # if cities are same return one else join with ,
    'kms': 'sum' # sums the kilometers of the duplicated rows

}
```

#### 5.1.4.4 Run Merging Operation

```python
# this takes about 10 seconds to run

# create a copy of 'duplicated_rows_DIC'
duplicated_rows_DIC = duplicated_rows_DIC.copy()

# replace 'nan' and '' in the 'history' column of 'duplicated_rows_DIC' DataFrame
duplicated_rows_DIC['history'].replace('nan', np.nan, inplace=True)
duplicated_rows_DIC['history'].replace('', np.nan, inplace=True)

def is_similar(a, b):
    return SequenceMatcher(None, a, b).ratio() > 0.6

# create an empty df to store the similar pairs
similar_histories = pd.DataFrame(columns=['ID1', 'ID2', 'Street1', 'Street2', 'History1', 'History2'])

for i in range(len(duplicated_rows_DIC)):
    for j in range(i+1, len(duplicated_rows_DIC)):
        history1 = duplicated_rows_DIC.iloc[i]['history']
        history2 = duplicated_rows_DIC.iloc[j]['history']
        if pd.notnull(history1) and pd.notnull(history2):
            if is_similar(history1, history2):
                new_row = pd.DataFrame({
                            'ID1':[duplicated_rows_DIC.iloc[i]['id']], 
                            'ID2':[duplicated_rows_DIC.iloc[j]['id']], 
                            'Street1':[duplicated_rows_DIC.iloc[i]['full street name']],
                            'Street2':[duplicated_rows_DIC.iloc[j]['full street name']],
                            'History1':[history1], 
                            'History2':[history2]
                            })
                similar_histories = pd.concat([similar_histories, new_row], ignore_index=True)

# temporarily remove max column width
pd.set_option('display.max_colwidth', None)
display(similar_histories)

# reset column width to default
pd.reset_option('display.max_colwidth')
```

### Gender Representation on Titles

<span style="font-family:Georgia;">The gender representation comparison is a crucial part of this project, since one of the goals is to shed light on the prevalence of gendered bias in street names. 

<span style="font-family:Georgia;">In this EDA, I have selected a few words that appear frequently in street names, as 'teacher' and 'doctor' - I'll make use that in portuguese almost all words are gendered; so it's possible to compare the representation of teachers broken down by gender.

```python

# create word pairs
word_pairs = [
    ('Professor', 'Professora'), # teacher
    ('Doutor', 'Doutura'), # doctor
    ('Dr','Dra'), # doctor abbreviation 
    ('Senhor', 'Senhora'), # mr / mrs
    ('Santo', 'Santa'), # saint 
    ('Padre', ['Irmã', 'Madre']) #priest and 'sister' - special case: summing the two because they're interchangeable
]

# initialize lists to store word counts
male_counts = []
female_counts = []

# convert street names to lowercase for comparison
DIC_database_clean_merged['street name'] = DIC_database_clean_merged['street name'].str.lower()

for pair in word_pairs:
    # handle special case
    if isinstance(pair[1], list):
        female_count = sum(DIC_database_clean_merged['street name'].str.contains(word.lower()).sum() for word in pair[1])
    else:
        female_count = DIC_database_clean_merged['street name'].str.contains(pair[1].lower()).sum()
    
    male_count = DIC_database_clean_merged['street name'].str.contains(pair[0].lower()).sum()

    male_counts.append(male_count)
    female_counts.append(female_count)

# create dataframe for plot
df = pd.DataFrame({'Male': male_counts, 'Female': female_counts}, index=[pair[0] for pair in word_pairs])

# create plot
colors = ['#555BED', '#FF71A4']
df.plot(kind='bar', figsize=(10,6), color=colors)
plt.title('Gender Comparison in Street Names')
plt.xlabel('Word')
plt.ylabel('Count')
plt.show()
```

### 5.4 First Names

<span style="font-family:Georgia;">The main use for a database with first names, is that when labelled by gender, later I can use it to create a little model that predicts when a name is more likely to be from a woman or a man. I'll scrape them individually from different sources, then I'll concatenate them together and remove the duplicates. 

```python
# WOMEN NAMES - WEBSITE 1/3

# url to be scraped
url = "https://palestraparaprofessores.com.br/familia/nomes-femininos/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0;Win64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}

# send request and start scraping
r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, 'html.parser')

# df that will become a full database
df = pd.DataFrame(columns=["women names"])

# finds the <p> tags 
p_tags = soup.find_all('p')

# row counter so I can remove some later by id
row = 0

# loop over the <p> tags
for tag in p_tags:
    # transforms <br> into line breaks
    for br in tag.findAll('br'):
        br.replace_with("\n")
        
    # split the text by the line breaks to get a list of names
    names = tag.text.split("\n")
    
    # adds each name into a row
    for name in names:
        df.loc[row] = name.strip()  # removes single white spaces
        row += 1
        
df = df.drop([0,1,2,3,36,57,58,81,112,137,164,322,416,439,482]) # removing some rows that were not names
df = df.reset_index(drop=True)
women_names_database = pd.DataFrame(df, columns=['women names'])
display(women_names_database)
```

```python
# WOMEN NAMES - WEBSITE 2/3

# url to be scraped
url = "https://4maos.com.br/nomes-antigos-femininos/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0;Win64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}

# send request and start scraping
r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, 'html.parser')

# df that will become a full database
df = pd.DataFrame(columns=["women names"])

# finds the <li> tags 
li_tags = soup.find_all('li')

# loop over the <li> tags
for li in li_tags:
    # finds the <strong> tag within the <li> tag
    strong_tag = li.find('strong')
    if strong_tag: # add to df
        new_df = pd.DataFrame({"women names": [strong_tag.text.strip()]})
        df = pd.concat([df, new_df], ignore_index=True)

df = df.drop([0,1,16]) # removing some rows that were not names
df = df.reset_index(drop=True)
women_names_database_2 = df.copy()

display(women_names_database_2.head())
```

```python
# WOMEN NAMES - WEBSITE 3/3

# url to be scraped
url = "https://www.significadodossonhosonline.com/nomes-antigos-femininos/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}

# send request and start scraping
r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, 'html.parser')

# df that will become a full database
df = pd.DataFrame(columns=["women names"])

# finds the <p> tags 
p_tags = soup.find_all('p')

# loop over the <p> tags
for p in p_tags:
    # finds the <strong> tags within the paragraphs
    strong_tags = p.find_all('strong')
    for tag in strong_tags:
        # get text from the <strong> tags and clean it
        name = tag.text.strip()
        name = re.sub(r'\d+\.', '', name)  # remove numbers followed by a dot aka '1.' 
        name = name.replace(':', '')  # remove colons
        name = name.strip()  # remove single spaces
        
        # if the name is not empty, add to the df
        if name:
            new_row = pd.DataFrame({"women names": [name]})
            df = pd.concat([df, new_row], ignore_index=True)

women_names_database_3 = pd.DataFrame(df, columns=['women names'])

display(women_names_database_3)
```

```python
# MEN NAMES - WEBSITE 2/2

# url to be scraped
url = "https://paisefilhos.uol.com.br/bebe/210-nomes-masculinos-para-bebe-ideias-fortes-e-lindas-para-voce-chamar-o-seu-filho/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}

# send request and start scraping
r = requests.get(url, headers=headers)
soup = BeautifulSoup(r.text, 'html.parser')

# df that will become a full database
df = pd.DataFrame(columns=["men names"])

# find the <h2> tags
h2_tags = soup.find_all('h2')

# loop over the <h2> tags
for h2 in h2_tags:
    # get text from the <strong> tags and clean it
    name = h2.text.strip()
    name = re.sub(r'\d+\.', '', name)  # remove numbers followed by a dot aka '1.' 
    name = name.replace(':', '')  # remove colons
    name = name.strip()  # remove single spaces

    # if the name is not empty, add to the df
    if name:
        new_df = pd.DataFrame({"men names": [name]})
        df = pd.concat([df, new_df], ignore_index=True)
        
men_names_database_2 = pd.DataFrame(df, columns=['men names'])
display(men_names_database_2)
```

```python
# concatenate all databases
men_names_database_combined = pd.concat([men_names_database, men_names_database_2])

# drop duplicated rows
men_names_database_combined = men_names_database_combined.drop_duplicates()

# reset index
men_names_database_combined = men_names_database_combined.reset_index(drop=True)

display(men_names_database_combined)
```
